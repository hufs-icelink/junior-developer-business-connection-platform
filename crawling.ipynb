```python
import re
```

```python
velog_page = requests.get(f"https://velog.io/@{velog_userId}")
soup = bs(velog_page.text, "html.parser")

links = soup.find_all('a')
link = []
for a in links:
    href = a.attrs['href']
    if f"{velog_userId}/" in href:
        link.append(href[1:])
        
print(set(link))

```
